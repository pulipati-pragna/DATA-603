1. Write Python code and use MapReduct to count occurrences of each word in the first text file (file1.txt). How many times each word is repeated?

```python
import re

# map function
def map_function(var):
    data_cleanup = ''.join([i for i in var.lower()])
    res = re.sub(r'[^\w\s]', '', data_cleanup)
    return [data_cleanup,1]

# reduce function
def reduce_function(key, val):
    word_count = [x[1] for x in val]
    return [key,sum(word_count)]

def reduce_function2(key):
    return key
```

```python
# seperating the words in the text file
file1 = open('file1.txt', encoding="utf8")
words1 = [word for line in file1 for word in line.split()]
print (words1)
```

```python
# using the map function for the words
map_function_result1 = map(map_function, words1)
```

```python
from itertools import groupby

# reducing the map function result
reduce_function_result1 = [];

for i, j in groupby(sorted(map_function_result1), key = lambda x: x[0]):
    reduce_function_result1.append(reduce_function(i, list(j)))
```

```python
print(reduce_function_result1)
```
